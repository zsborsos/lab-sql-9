```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import getpass
password = getpass.getpass()
```

Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
data_may = pd.read_sql_query('select * from sakila.rentals_may;', engine)
data_june = pd.read_sql_query('select * from sakila.rentals_june;', engine)
```

Check the number of rentals for each customer for May

```python
df_may = pd.DataFrame(data_may)
```

```python
df_may.head()
```

```python
engine = create_engine(connection_string)
rentals_may = pd.read_sql_query('SELECT customer_id, count(rental_id) AS May FROM sakila.rentals_may GROUP BY customer_id ORDER BY customer_id', engine)
rentals_may
```

```python
df_rentals_may = pd.DataFrame(rentals_may)
df_rentals_may
```

Check the number of rentals for each customer for June

```python
df_june = pd.DataFrame(data_june)
```

```python
df_june.head()
```

```python
engine = create_engine(connection_string)
rentals_june = pd.read_sql_query('SELECT customer_id, count(rental_id) AS June FROM sakila.rentals_june GROUP BY customer_id ORDER BY customer_id', engine)
rentals_june
```

```python
df_rentals_june = pd.DataFrame(rentals_june)
df_rentals_june
```

Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
df_rentals_may_june = pd.merge(df_rentals_may, df_rentals_june, how='outer', on='customer_id')
```

```python
df_rentals_may_june
```

```python
import numpy as np
rules = [
    (df_rentals_may_june['May'] > df_rentals_may_june['June']), (df_rentals_may_june['May'] < df_rentals_may_june['June']), (df_rentals_may_june['May']== df_rentals_may_june['June']), df_rentals_may_june['May'].isnull(), df_rentals_may_june['June'].isnull()]

options = ['Borrowed more films in May', 'Borrowed more films in June', 'Borrowed the same number of films in May and June', 'Only borrowed films in June', 'Only borrowed films in May']

df_rentals_may_june['Comparison'] = np.select(rules, options)
df_rentals_may_june
```
